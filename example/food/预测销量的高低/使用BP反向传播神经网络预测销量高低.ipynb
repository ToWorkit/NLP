### 建立的神经网络有3个输入节点，10个隐藏节点和1个输出节点_p98

```python
import pandas as pd
```

```python
# 数据导入
inputfile = 'E:\\NLP\\Python数据分析与挖掘实战\\chapter5\\demo\\data\\sales_data.xls'
data = pd.read_excel(inputfile, index_col = u'序号')
# print(data)
print(type(data.iloc[:, 3]))
```

```python
# 转换数据
# 1 -> 好，是，高。0 -> 坏，否，低
data[data == u'好'] = 1
data[data == u'是'] = 1
data[data == u'高'] = 1
data[data != 1] = 0
# 取 3 列(因为以 序号 做了索引列 所以不含序号)，并转化为 numpy矩阵形式，再将数值类型转为 int
x = data.iloc[:,:3].as_matrix().astype(int)
# 取 下标为 3 的那一列数据,同样不含序号
y = data.iloc[:, 3].as_matrix().astype(int)
```

```python
# 神经网络算法库 Keras
from keras.models import Sequential
# 激活函数
from keras.layers.core import Dense, Activation
```

```python
# 建立模型
model = Sequential()
# 三个输入节点，十个隐藏节点
model.add(Dense(input_dim = 3, units = 10))
# 用 relu函数 作为激活函数，可以大幅提高准确度
model.add(Activation('relu'))
# 十个隐藏节点(就是上面的)，一个输出节点
model.add(Dense(input_dim = 10, units = 1))
# 因为是 0-1 输出，所以使用 sigmoid函数 作为激活函数 -> p_96 非线性转移函数
# https://baike.baidu.com/item/Sigmoid%E5%87%BD%E6%95%B0/7981407?fr=aladdin
model.add(Activation('sigmoid'))
```

```python
# 编译模型，因为是二元分类，所以指定损失函数为 binary_crossentropy
# 其他常见的损失函数还有 mean_squared_error, categorical_crossentropy等
# 求解方法（迭代优化）指定 adam，其他的还有 sgd(随机梯度下降法)， rmsprop等
model.compile(loss = 'binary_crossentropy', optimizer = 'adam')
```

```python
# 训练模型，学习 1000次
model.fit(x, y, epochs = 1400, batch_size = 10)
# 分类预测
yp = model.predict_classes(x).reshape(len(y))
```
